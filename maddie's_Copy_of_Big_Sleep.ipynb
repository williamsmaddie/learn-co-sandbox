```
terminate = False
 
def signal_handling(signum,frame):
    global terminate
    terminate = True
 
for attempt in range(num_attempts):
  print("attempt number ",attempt,'\n')
 
  model = Imagine(
      text = TEXT,
      save_every = SAVE_EVERY,
      lr = LEARNING_RATE,
      iterations = ITERATIONS,
      epochs = EPOCHS,
      save_progress = SAVE_PROGRESS,
      save_best = SAVE_BEST,
      seed = SEED,
      text_min = TEXT_MIN
      # resolution = RESOLUTION
  )
 
# faster way?
  for epoch in trange(model.epochs, desc = 'epochs'):
    for i in trange(model.iterations, desc = 'iteration'):
        model.train_step(epoch, i)
 
 # more verbose way - slower?
  # model()
 
  filename = model.text_path + "-attempt-" + str(attempt) + "-" + str(datetime.now()) +".png"
  shutil.copy(model.text_path + ".best.png", filename)
 
  #saving and displaying attempt
  print('\n')
  print("filename is: ", filename)
 
    # print attempt
  image=Image(filename)
  display(image)
 
  # save image
  if gauth.access_token_expired:
    print("fyi token expired but i refreshed it")
    gauth.Refresh()
 
  uploaded = drive.CreateFile({'title': filename})
 
  uploaded.SetContentFile(filename)
  uploaded.Upload()
  # print('Uploaded file with ID {}'.format(uploaded.get('id')))
  # print('Uploaded file with ID {}'.format(img_id))
  print('File name is ', filename)
 
 
 
  #cleaning and trying again
  try:
      time.sleep(2)
      del model
      time.sleep(2)
      torch.cuda.empty_cache()
  except Exception:
      torch.cuda.empty_cache()
      torch.cuda.empty_cache()
  except Exception:
      torch.cuda.empty_cache()
```
